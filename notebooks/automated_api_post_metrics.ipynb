# Results from Automated Metrics

```python
# SET UP
import pandas as pd
from datetime import datetime
from notebook_to_lib_config import set_path 
config = set_path()
POST_METRICS_CSV = config["POST_METRICS_PATH"]
M_INTITAL_EXTRACT = config["M_INTITAL_EXTRACT"]
CLEANED_DATA_PATH = config["CLEANED_DATA_PATH"]
```

```python
post_metrics = pd.read_csv(POST_METRICS_CSV)
length = len(post_metrics)
print(f"{length} rows of data pulled from Instagram Graph API:")
post_metrics.tail()
```

```python
post_metrics['Post ID'] = post_metrics['id']
intital_extract = pd.read_csv(M_INTITAL_EXTRACT)
merged_df = pd.merge(post_metrics, intital_extract, on='Post ID', how='left')
print("Data merged with initial extract: ")
merged_df.head()
```

```python
merged_df.columns
```

```python
length = len(merged_df['Post ID'].value_counts())
print(f"Number of posts gathered: {length}")
print(merged_df['Post type'].value_counts())
```

```python
# Export to Excel
merged_df.to_csv(CLEANED_DATA_PATH + "daily_post_metrics.csv", index=False) 
merged_df.to_excel(CLEANED_DATA_PATH + "daily_post_metrics.xlsx", sheet_name="post_metrics", index=False) 
```
